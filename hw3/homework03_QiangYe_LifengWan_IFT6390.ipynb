1. Show that $\text{sigmoid}(x) = \frac{1}{2}(\text{tanh}(\frac{1}{2}x)+1)$  
**Answer**   
\begin{aligned}
\frac{1}{2}\bigl(\text{tanh}(\frac{1}{2}x) + 1\bigr) &= \frac{1}{2}\bigl(\frac{exp(\frac{1}{2}x) - exp(-\frac{1}{2}x)}{exp(\frac{1}{2}x) + exp(-\frac{1}{2}x)} + 1 \bigr)\\
&= \frac{1}{2}\bigl(\frac{2\ exp(\frac{1}{2}x)}{exp(\frac{1}{2}x) + exp(-\frac{1}{2}x)}\bigr)\\
&= \frac{exp(\frac{1}{2}x)}{exp(\frac{1}{2}x) + exp(-\frac{1}{2}x)}\\
&= \frac{1}{1 + \frac{exp(-\frac{1}{2}x)}{exp(\frac{1}{2}x)}}\\
&= \frac{1}{1+exp(-\frac{1}{2}x - \frac{1}{2}x)}\\
&= \frac{1}{1+exp(-x)} \\
&= \text{sigmoid}(x)\\
\end{aligned}<br>  

2. Show that $ln\ \text{sigmoid}(x) =  -\text{softplus}(-x)$  
**Answer**  
\begin{aligned}
ln\ \text{sigmoid}(x) &= ln\ \frac{1}{1+exp(-x)}\\
&= -ln\ (1+exp(-x))\\
&= -\text{softplus}(-x)\\
\end{aligned}<br>  

3. Show that the derivative of the sigmoid is: $\text{sigmoid}'(x) = \frac{d\ \text{sigmoid}}{d\ x}(x) = \text{sigmoid}(x)(1-\text{sigmoid}(x))$  
**Answer**
\begin{aligned}
\text{sigmoid}'(x) &= \frac{d\ \text{sigmoid}}{d\ x}(x)\\
&= \frac{d\ \frac{1}{1+exp(-x)} }{dx}\\
&= -(\frac{1}{1+exp(-x)})^2 \frac{d\ (1+exp(-x))}{dx}\\
&= \frac{1}{(1+exp(-x))^2} exp(-x)\\
&= \frac{1}{1+exp(-x)} \frac{exp(-x)}{1+exp(-x)}\\
&= \frac{1}{1+exp(-x)} \frac{1 + exp(-x) - 1}{1+exp(-x)}\\
&= \frac{1}{1+exp(-x)} \Bigl(1 - \frac{1}{1+exp(-x)}\Bigr)\\
&= \text{sigmoid}(x)(1-\text{sigmoid}(x))
\end{aligned}<br>  

4. Show that the tanh derivative is: tanh'($x$) = 1 - tanh$^2(x)$  
**Answer**  
\begin{aligned}
tanh'(x) &= \frac{d\ (\frac{exp(x)-exp(-x)}{exp(x)+exp(-x)})}{d\ x}\\ 
&= \frac{(exp(x)+exp(-x))(exp(x)-exp(-x))'-(exp(x)-exp(-x))(exp(x)+exp(-x))'}{(exp(x)+exp(-x))^2} \\
&= \frac{(exp(x)+exp(-x))^2 - (exp(x)-exp(-x))^2}{(exp(x)+exp(-x))^2}\\
&= 1 - \frac{(exp(x)-exp(-x))^2}{(exp(x)+exp(-x))^2}\\
&= 1 - \text{tanh}^2(x)\\
\end{aligned}<br>

We will now consider a <b>regularized</b> empirical risk: $\tilde R = \hat R + \mathcal L(\theta)$, where $\theta$ is the vector of all the parameters in the network and $\mathcal L(\theta)$ describes a scalar penalty as a function of the parameters $\theta$. The penalty is given importance according to a prior preferences for the values of $\theta$. The $L_2$(quadratic) regularization that penalizes the square norm (norm $L_2$) of the weights (but not the biases) is more standard, is used in ridge regresion and is sometimes called "weight-decay". Here we shall consider a double regularization $L_2$ and $L_1$ which is sometimes named "elastic net" and we will use different <b>hyperparameters</b> (positive scalars $\lambda_{11}, \lambda_{12}, \lambda_{21}, \lambda_{22}$) to control the effect of the regularization at each layer  
\begin{aligned}
\mathcal L(\theta) &= \mathcal L(\mathbf W^{(1)}, \mathbf b^{(1)}, \mathbf W^{(2)}, \mathbf b^{(2)})\\
&= \lambda_{11} \Vert \mathbf W^{(1)} \Vert_1 +
\lambda_{12} \Vert \mathbf W^{(1)} \Vert_2^2 +
\lambda_{21} \Vert \mathbf W^{(2)} \Vert_1 +
\lambda_{22} \Vert \mathbf W^{(2)} \Vert_2^2 \\
&= \lambda_{11}\Bigl( \sum_{i,j} \lvert \mathbf W^{(1)}_{ij} \rvert \Bigr) +
\lambda_{12}\Bigl( \sum_{i,j} ( \mathbf W^{(1)}_{ij} )^2 \Bigr) +
\lambda_{21}\Bigl( \sum_{i,j} \lvert \mathbf W^{(2)}_{ij} \rvert \Bigr) +
\lambda_{22}\Bigl( \sum_{i,j} ( \mathbf W^{(2)}_{ij} )^2 \Bigr) 
\end{aligned}  
We will in fact minimize the regularized risk $\tilde R$ instead of $\hat R$. How does this change the gradient with respect to the different parameters?  
**Answer**  
$$
\frac{\partial \tilde R}{\partial \mathbf W^{(1)}_{ij}} = 
\frac{\partial \hat R}{\partial \mathbf W^{(1)}_{ij}} + 
\lambda_{11} \text{sign}(\mathbf W^{(1)}_{ij}) + 2\lambda_{12} \mathbf W^{(1)}_{ij}
$$
$$
\frac{\partial \tilde R}{\partial \mathbf W^{(2)}_{ij}} = 
\frac{\partial \hat R}{\partial \mathbf W^{(2)}_{ij}} + 
\lambda_{21} \text{sign}(\mathbf W^{(2)}_{ij}) + 2\lambda_{22} \mathbf W^{(2)}_{ij}
$$
$$\frac{\partial \tilde R}{\partial \mathbf b^{(2)}_{i}} = 
\frac{\partial \hat R}{\partial \mathbf b^{(2)}_{i}}
$$
$$
\frac{\partial \tilde R}{\partial \mathbf b^{(1)}_{i}} = 
\frac{\partial \hat R}{\partial \mathbf b^{(1)}_{i}},
$$

### 2. Implementation of MLP class

Note: 
1. Initialization of the network parameters(`W1, W2, b1, b2`) is not performed in the function `__init__` as we might not know the number of features of input and output until we know the X and y.
2. Two set of methods (loop and matrix expression) about the forward and backward propagation are implemented as required by the experiments which will be explained in detail in the answer of related experiments.

The completed implementation of class `MLP` is as follows:

````python
class MLP():
    """one hidden layer with RELU and Softmax
    """
    def __init__(self, 
                 n_hidden = 2,
                 lambdas = [0, 0, 0, 0],
                 learning_rate = 1e-3,
                 epochs = 200,
                 batch_size = 10,
                 verbose = False,
                 n_input = None,
                 n_output = None,
                 early_stop = False
                ):
        self.n_i = n_input
        self.n_o = n_output
        self.n_h = n_hidden
        self.lambdas = lambdas
        self.lr = learning_rate
        self.epochs = epochs
        self.batch_size = batch_size
        self.early_stop = early_stop
        self.verbose = verbose
        

    def _init_weights_biases(self, X, y, n_output = None):
        """number of input and output neurons are decided when dataset
        is known, initialization of parameters are performed here.
        Params
            X: input of training data ndarray (sample_size, n_features)
            y: true class labels of X ndarray (sample_size, 1)
            n_output: number of true classes, is None, can be calculated 
        from y.
        Returns
            None
        """
        if n_output is None:
            y = y.reshape(-1, )
            self.n_o = int(np.max(y) + 1)
        else:
            self.n_o = n_output
        np.random.seed(0) # for compare
        sample_size, n_input = X.shape
        self.n_i = n_input
        _high = 1 / np.sqrt(self.n_i)
        self.W1 = np.random.uniform(-1 * _high, _high, 
                                    (self.n_h, self.n_i))
        self.b1 = np.zeros((self.n_h, 1)) # column vector
        
        _high = 1 / np.sqrt(self.n_h)
        self.W2 = np.random.uniform(-1 * _high, _high, 
                                    (self.n_o, self.n_h))
        self.b2 = np.zeros((self.n_o, 1)) # column vector


    def _init_grads(self):
        """Initialization of gradient of parameters
        Params:None
        Returns
            grads: zero gradient of all parameters dict
        """
        grad_W1 = np.zeros_like(self.W1)
        grad_W2 = np.zeros_like(self.W2)
        grad_b1 = np.zeros_like(self.b1)
        grad_b2 = np.zeros_like(self.b2)
        grads = {"grad_W1":grad_W1, "grad_W2":grad_W2, 
                 "grad_b1":grad_b1, "grad_b2":grad_b2,
                }
        return grads
    
    
    def _fprop_one_example(self, x):
        """forward propagation of the network using one example.
        Params
            x: example ndarray (n_features, 1)
        Returns
            cache: a dict variable including computation results used for 
        back propagation.
        """
        x = x.reshape(-1, 1) # column vector (n_input, 1)
        ha = np.dot(self.W1, x) + self.b1 # (n_hidden, 1)
        hs = np.maximum(ha, 0) # (n_hidden, 1)
        
        oa = np.dot(self.W2, hs) + self.b2 # (n_output, 1)
        os = my_softmax(oa.T).T # (n_output, 1) softmax
        
        cache = {"ha": ha, "hs": hs, "oa": oa, "os": os, "x":x}
        return cache
    
        
    def _bprop_one_example(self, cache, y):
        """backward propagation of the network using one example.
        Params
            cache: a dict variable from the output of forward propagation.
            y: int, true class index of the example
        Returns
            grads: gradient of the parameters of the network.
        """
        ha = cache["ha"] # (n_hidden, 1)
        hs = cache["hs"] # (n_hidden, 1)
        oa = cache["oa"] # (n_output, 1)
        os = cache["os"] # (n_output, 1)
        x = cache["x"]   # (n_input, 1)
        grad_oa = np.copy(os) # (n_output, 1)
        np.put(grad_oa, y, grad_oa[y,0]-1)
        
        grad_W2 = np.dot(grad_oa, hs.T) # (n_output, n_hidden)
        grad_b2 = grad_oa # (n_output, 1)
        
        grad_hs = np.dot(self.W2.T, grad_oa) # (n_hidden, 1)
        grad_ha = np.multiply(grad_hs, np.maximum(np.sign(ha),0)) 
        # (n_hidden, 1)
        
        grad_W1 = np.dot(grad_ha, x.T) # (n_hidden, n_input)
        grad_b1 = grad_ha # (n_hidden, 1)
        
        # grad_x = np.dot(self.W1.T, grad_ha) # (n_input, 1)
        lambdas = self.lambdas
        if not lambdas[0] == 0:
            grad_W1 += lambdas[0] * np.sign(self.W1)
        if not lambdas[1] == 0:
            grad_W1 += 2 * lambdas[1] * self.W1
        if not lambdas[2] == 0:
            grad_W2 += lambdas[2] * np.sign(self.W2)
        if not lambdas[3] == 0:
            grad_W2 += 2 * lambdas[3] * self.W2
            
        grads = {"grad_W1":grad_W1, "grad_W2":grad_W2, 
                 "grad_b1":grad_b1, "grad_b2":grad_b2,
                 #"grad_hs":grad_hs, "grad_ha":grad_ha,
                }
        return grads
    

    def fit_non_matrix_expression(self, X, y):
        """fit X to y without using matrix expression. parameters
        are updated after every batch_size. In each batch_size, 
        grads are accumulated by a looping over each example.
        Params
            X: training X ndarray (batch_size, n_features)
            y: training labels ndarray (batch_size, )
        Returns
            losses: list of loss of each epoch [float]
            errors: list of error rate of each epoch [float]
        """
        self._init_weights_biases(X, y)
        
        sample_size, _ = X.shape
        batches = int(np.ceil(sample_size / self.batch_size))
        losses, errors = [], [] # store loss and error each epoch
        for epoch in range(self.epochs):
            loss, err = 0.0, 0.0
            for j in range(batches):
                b_start = j * self.batch_size
                b_end = min(sample_size, (j+1) * self.batch_size)
                batch_X = X[b_start:b_end, :]
                batch_y = y[b_start:b_end]
                grads = self._init_grads()
                for k in range(b_end - b_start):
                    cache = self._fprop_one_example(batch_X[k])
                    grads_tmp = self._bprop_one_example(cache, batch_y[k])
                    for key in grads:
                        # compute mean grads[key]
                        grads[key] += (grads_tmp[key] - grads[key])/(k + 1)
                    loss += self._compute_one_loss(cache["os"], batch_y[k])
                self._update_params(grads)
            # end of mini_batch

            error = self.compute_error(self.predict(X), y)
            errors.append(error)
            loss /= sample_size 
            loss += self._compute_regular_loss()
            losses.append(loss)
        # end of one epoch
        return losses, errors
        
    
    def _fprop(self, X):
        """
        forward propagation using Matrix expression
        params
            X: input data (batch_size, n_input)
        returns
            os: output of the network (batch_size, n_output)
        """
        ha = np.dot(X, self.W1.T) + self.b1.T # (batch_size, n_hidden)
        hs = np.maximum(ha, 0) # relu (batch_size, n_hidden)
        
        oa = np.dot(hs, self.W2.T) + self.b2.T # (batch_size, n_output)
        os = my_softmax(oa) # (batch_size, n_output)
        
        cache = {"ha": ha, "hs": hs, "oa": oa, "os": os, "X":X}
        return cache
    

    def _bprop(self, cache, y):
        """
        backward propagation using Matrix expression
        params
            os: output of network (batch_size, n_output)
            y: true label class (batch_size, 1)
        returns:
            grads: gradients( grad_W1, grad_W2, grad_b1, grad_b2)
        """
        ha = cache["ha"] # (batch_size, n_hidden)
        hs = cache["hs"] # (batch_size, n_hidden)
        oa = cache["oa"] # (batch_size, n_output)
        os = cache["os"] # (batch_size, n_output)
        X = cache["X"]   # (batch_size, n_input)
        
        batch_size, n_o = os.shape # (batch_size, n_output)
        grad_oa = os - one_hot(n_o, y) #(batch_size, n_output)
        # grad_oa = np.mean(grad_oa, axis = 0, keepdims = True).T
        # (n_output, 1)
        
        grad_W2 = np.dot(grad_oa.T, hs) # (n_output, n_hidden)
        grad_b2 = grad_oa # (batch_size, n_output)
        
        grad_hs = np.dot(grad_oa, self.W2) # (batch_size, n_hidden)
        grad_ha = np.multiply(grad_hs, np.maximum(np.sign(ha),0)) 
        # (batch_size, n_hidden)
        
        grad_W1 = np.dot(grad_ha.T, X) # (n_hidden, n_input)
        grad_b1 = grad_ha # (batch_size, n_hidden)
        
        # grad_x = np.dot(grad_ha, self.W1) # (batch_size, d)
        
        # mean grad
        grad_W2 /= batch_size
        grad_W1 /= batch_size
        grad_b2 = np.mean(grad_b2, axis = 0, keepdims = True).T 
        grad_b1 = np.mean(grad_b1, axis = 0, keepdims = True).T 
        
        lambdas = self.lambdas
        if not lambdas[0] == 0:
            grad_W1 += lambdas[0] * np.sign(self.W1)
        if not lambdas[1] == 0:
            grad_W1 += 2 * lambdas[1] * self.W1
        if not lambdas[2] == 0:
            grad_W2 += lambdas[2] * np.sign(self.W2)
        if not lambdas[3] == 0:
            grad_W2 += 2 * lambdas[3] * self.W2
        
        grads = {"grad_W1":grad_W1, "grad_W2":grad_W2, 
                 "grad_b1":grad_b1, "grad_b2":grad_b2,
                 #"grad_hs":grad_hs, "grad_ha":grad_ha,
                }
        return grads
    
    
    def fit(self, X, y):
        """fit X to y. Matrix operation
        Params
            X: training X ndarray (sample_size, n_features)
            y: training labels ndarray (sample_size, )
        Returns
            losses: list of loss of each epoch [float]
            errors: list of error rate of each epoch [float]
        """
        self._init_weights_biases(X, y)
        sample_size, _ = X.shape
        batches = int(np.ceil(sample_size / self.batch_size))
        losses, errors = [], []
    
        for epoch in range(self.epochs):
            for j in range(batches):
                b_start = j * self.batch_size
                b_end = min(sample_size, (j+1) * self.batch_size)
                batch_X = X[b_start:b_end, :]
                batch_y = y[b_start:b_end]
                cache = self._fprop(batch_X)
                grads = self._bprop(cache, batch_y)
                self._update_params(grads)
            # end batch_size
            os = self._fprop(X)["os"]
            loss = self.compute_empirical_risk(os, y)
            losses.append(loss)
            
            oy = np.argmax(os, axis = 1)
            error = self.compute_error(oy, y)
            errors.append(error)
        # end epoch
        return losses, errors
    
    
    def predict_probs(self, X):
        """predict probabilities (os) of given data
        Params
            X: examples. ndarray (certain_size, n_input)
        Returns
            os: probabilities of X belong to every class. 
        ndarray (certain_size, n_output)
        """
        return self._fprop(X)["os"]
    

    def predict(self, X):
        """predict class labels of given data
        Params
            X: examples ndarray (certain_size, n_input)
        Returns
            y_output: predicted label of given data. ndarray 
        (certain_size, )
        """
        os = self.predict_probs(X)
        return np.argmax(os, axis = 1)   
    
              
    def _backup_params(self):
        """save current network parameters
        Params: None
        Returns:
            parameters saved. dict.
        """
        old_params = {"W1":np.copy(self.W1),
                      "W2":np.copy(self.W2),
                      "b1":np.copy(self.b1),
                      "b2":np.copy(self.b2)
                          }
        return old_params
    
    
    def _restore_params(self, old_params = None):
        """restore network parameters by change network's parameters
        Params
            old_params: parameters restored
        Returns:
            None
        """
        if old_params is None:
            return
        self.W1 = old_params["W1"]
        self.W2 = old_params["W2"]
        self.b1 = old_params["b1"]
        self.b2 = old_params["b2"]
        
        
    def _update_params(self, grads, learnig_rate = None):
        """update network parameters using given gradients and larning 
        rate
        Params
            grads: gradients of parameters. dict
            learning_rate: learning rate. float
        Returns
            None
        """
        lr = learnig_rate
        if lr is None:
            lr = self.lr
        self.W1 -= lr * grads["grad_W1"]
        self.W2 -= lr * grads["grad_W2"]
        self.b1 -= lr * grads["grad_b1"]
        self.b2 -= lr * grads["grad_b2"]

        
    def compute_error(self, oy, y):
        """compute mis classification rate 
        Params
            oy: output of a classifier ndarray (size, )
            y: true class labels ndarray (size, )
        Returns
            error: misclassified error. float
        """
        return  float(1 - np.sum(oy == y) / len(oy))
    
    
    def _compute_one_loss(self, os, y):
        """loss of one example
        Params
            os: column vector output (n_output, 1)
            y: scalar int
        Returns loss of one example. float
        """
        os = os.reshape(-1, 1)
        os_y = os[y, 0] + 1e-100 # avoid np.log(0)
        return float(-np.log(os_y))
    
    
    def _compute_regular_loss(self):
        """regular loss
        Params 
            None
        Returns
            None
        """
        loss = 0.0
        lambdas = self.lambdas
        if not lambdas[0] == 0:
              loss += lambdas[0] * np.sum(np.abs(self.W1))
        if not lambdas[1] == 0:
              loss += lambdas[1] * np.sum(np.power(self.W1, 2))
        if not lambdas[2] == 0:
              loss += lambdas[2] * np.sum(np.abs(self.W2))
        if not lambdas[3] == 0:
              loss += lambdas[3] * np.sum(np.power(self.W2, 2)) 
        return float(loss)
        
        
    def compute_empirical_risk(self, os, y):
        """
        compute $\tilde R = \hat R + \mathcal L(\theta)$
        params
            os: output of network (sample_size, n_output)
            y: true label (sample_size, class_index)
        """
        batch_size, n_output = os.shape
        y = y.reshape(-1, )
        y = y.astype(int)
        os_y = os[np.arange(batch_size), y]
        loss = np.mean(-np.log(os_y)) # might cause warning from np.log()
        loss += self._compute_regular_loss()
        return float(loss)
````

        x = cache["x"]   # (n_input, 1)
        grad_oa = np.copy(os)
        np.put(grad_oa, y, grad_oa[y,0]-1)
        
        grad_W2 = np.dot(grad_oa, hs.T) # (n_output, n_hidden)
        grad_b2 = grad_oa # (n_output, 1)
        
        grad_hs = np.dot(self.W2.T, grad_oa) # (n_hidden, 1)
        grad_ha = np.multiply(grad_hs, np.maximum(np.sign(ha),0)) # (n_hidden, 1)
        
        grad_W1 = np.dot(grad_ha, x.T) # (n_hidden, n_input)
        grad_b1 = grad_ha # (n_hidden, 1)
        
        # grad_x = np.dot(self.W1.T, grad_ha) # (n_input, 1)
        lambdas = self.lambdas
        if not lambdas[0] == 0:
            grad_W1 += lambdas[0] * np.sign(self.W1)
        if not lambdas[1] == 0:
            grad_W1 += 2 * lambdas[1] * self.W1
        if not lambdas[2] == 0:
            grad_W2 += lambdas[2] * np.sign(self.W2)
        if not lambdas[3] == 0:
            grad_W2 += 2 * lambdas[3] * self.W2
            
        grads = {"grad_W1":grad_W1, "grad_W2":grad_W2, 
                 "grad_b1":grad_b1, "grad_b2":grad_b2,
                 #"grad_hs":grad_hs, "grad_ha":grad_ha,
                }
        return grads

Two other functions `is_grad_equal` and `is_grads_equal` are implemented to check if the grads by finite difference method and by back propagation of the network are equal. Only when two values of the gradient with respect to each parameter (`W1, W2, b1, b2`)  from two methods are almost equal (difference between 0.01), will the functions give a `True` result and print "All grads are equal".

See the following codes for details.

````python
def grad_finite_diff_one_example(mlp, x, c, epsilon = 1e-5):
    grads = mlp._init_grads()
    params = {"grad_W1":mlp.W1, "grad_W2":mlp.W2, 
              "grad_b1":mlp.b1, "grad_b2":mlp.b2,
             }
    for grad_key in grads:
        grad = grads[grad_key]
        param = params[grad_key]
        m, n = grad.shape
        for i in range(m):
            for j in range(n):
                old_value = param[i, j]

                value_plus = old_value + epsilon
                param[i, j] = value_plus
                cache = mlp._fprop_one_example(x)
                loss_plus = mlp._compute_one_loss(cache["os"], c)
                loss_plus += mlp._compute_regular_loss()
                
                value_minor = old_value - epsilon
                param[i, j] = value_minor
                cache = mlp._fprop_one_example(x)
                loss_minor = mlp._compute_one_loss(cache["os"], c)
                loss_minor += mlp._compute_regular_loss()
                
                grad[i, j] = (loss_plus - loss_minor)/(2.0 * epsilon)
                param[i, j] = old_value
    return grads


def is_grad_equal(grad1, grad2, epsilon = 1e-6):
    """check if all element values of two grad array are equal"""
    if not grad1.shape == grad2.shape:
        return False
    m, n = grad1.shape
    for i in range(m):
        for j in range(n):
            if abs(grad1[i,j] - grad2[i,j]) > epsilon:
                return False
    return True


def is_grads_equal(grads1, grads2, epsilon = 1e-6):
    for grad_key in grads1:
        if not is_grad_equal(grads1[grad_key], grads2[grad_key], epsilon):
            print("Failure in gradient check of param: {}".format(grad_key))
            return False
    print("All gradients are equal")
    return True
````

### 3. 
Add a hyperparameter for the minibatch size K to allow compute the gradients on a minibatch of $K$ examples (in a matrix), by looping over the $K$ examples (this is a small addition to your previous code).

**Answer / Code**

In this experiment, two member function of class `MLP` were used:

    _compute_regular_loss(self)
    
The loss of one example is calculated by the following equation:
$$L(\mathbf x, y) = -log\ \mathbf o^s_y(\mathbf x)$$
which we have already seen in question 2.4 of theoretical part. The corresponding codes are:

    def _compute_one_loss(self, os, y):
        """loss of one example
        params
            os: column vector output (n_output, 1)
            y: scalar int
        """
        os = os.reshape(-1, 1)
        os_y = os[y, 0] + 1e-100 # avoid np.log(0)
        return float(-np.log(os_y))

The loss of regularization is also a part of the loss on either one example or batch examples, and they can be derived by following codes:

    def _compute_regular_loss(self):
        loss = 0.0
        lambdas = self.lambdas
        if not lambdas[0] == 0:
              loss += lambdas[0] * np.sum(np.abs(self.W1))
        if not lambdas[1] == 0:
              loss += lambdas[1] * np.sum(np.power(self.W1, 2))
        if not lambdas[2] == 0:
              loss += lambdas[2] * np.sum(np.abs(self.W2))
        if not lambdas[3] == 0:
              loss += lambdas[3] * np.sum(np.power(self.W2, 2)) 
        return float(loss)
        

For the method of back propagation, the gradient with respect to a parameter on a batch size of $K$ examples can be considered as the mean of the gradient on each sample in the batch, which is implemented in the function: `grad_back_prop_batch(mlp, X, y)`.

For the method of finite difference method, the gradient relies on the tiny change of average loss by the tiny change of the parameter. So, the key is to compute the average loss on the minibatch. See the implementation of the function `grad_finite_diff_batch(mlp, X, y, epsilon = 1e-5)` for details.

````python
def grad_back_prop_batch(mlp, X, y):
    batch_size, n_input = X.shape
    y = y.reshape(-1, )
    grads1 = mlp._init_grads()
    for i in range(batch_size):
        x, c = X[i,:], y[i]
        cache = mlp._fprop_one_example(x)
        grads = mlp._bprop_one_example(cache, c)
        for key in grads1:
            # compute mean grads1[key]
            grads1[key] += (grads[key] - grads1[key])/(i + 1)
    return grads1        
        
        
def grad_finite_diff_batch(mlp, X, y, epsilon = 1e-5):
    grads = mlp._init_grads()
    params = {"grad_W1":mlp.W1, "grad_W2":mlp.W2, 
              "grad_b1":mlp.b1, "grad_b2":mlp.b2,
             }
    batch_size, _ = X.shape
    y = y.reshape(-1, )
    y = y.astype(int)
    for grad_key in grads:
        grad = grads[grad_key]
        param = params[grad_key]
        m, n = grad.shape
        for i in range(m):
            for j in range(n):
                old_value = param[i, j]

                value_plus = old_value + epsilon
                param[i, j] = value_plus
                loss_plus = 0
                for k in range(batch_size):
                    cache = mlp._fprop_one_example(X[k,:])
                    loss_plus += mlp._compute_one_loss(cache["os"], y[k])
                loss_plus /= batch_size
                loss_plus += mlp._compute_regular_loss()

                
                value_minor = old_value - epsilon
                param[i, j] = value_minor
                loss_minor = 0
                for k in range(batch_size):
                    cache = mlp._fprop_one_example(X[k,:])
                    loss_minor += mlp._compute_one_loss(cache["os"], y[k])
                loss_minor /= batch_size
                loss_minor += mlp._compute_regular_loss()
                grad[i, j] = (loss_plus - loss_minor)/(2.0 * epsilon)
                param[i, j] = old_value
    return grads
````

The graients from two methods should be equal. Again, we can check them by the functions used in the experiment 1. 

Luckly, We got the same gradients.

````python
# create a mlp with certain neurons in hidden layer
mlp = MLP(n_hidden = 2)
mlp._init_weights_biases(X, y) # initialize weights and bias
batch_size = 10
start = np.random.randint(0, len(X)-batch_size)
batch_X = X[start:start + batch_size,:]
batch_y = y[start:start + batch_size]

grads1 = grad_back_prop_batch(mlp, batch_X, batch_y)
grads2 = grad_finite_diff_batch(mlp, batch_X, batch_y)
is_grads_equal(grads1, grads2)
````

**Answer / Code**

We specifically implemented the following method to train our network on different hyperparameters and plot the corresponding decision regions.

As to early stopping, we set the tolerant epoch times `n_patience = 20`. When error rate on validate dataset starts to increase, we observe the consequesnt `n_patience` number of epochs before stopping training.

Several member functions of class `MLP` are used in this experiments:

The purposes of these functions are just as simple as the names of the functions tell, and they are also easy to implement. No further explanation for these functions. See the implementation of the class `MLP` for details.

````python
def train_by_looping(train_X, train_y, valid_X, valid_y,
          n_hidden = 8, early_stop = False, weight_decay = [0.0, 0.0, 0.0, 0.0],
          learning_rate = 1e-3, epochs = 1000, batch_size = 10
         ):
    mlp = MLP(n_hidden = n_hidden,
              learning_rate = learning_rate,
              epochs = epochs,
              batch_size = batch_size,
              early_stop = early_stop,
              lambdas = weight_decay)
    
    mlp._init_weights_biases(train_X, train_y)
    
    # for early stopping
    best_params = None # params that performs best in valid set
    n_patientce = 200 # maximun numbers of epoch tolerated when mlp
                      # with current params doesn't perform better than
                      # with best_params.
    n_epochs_bad = 0 # epochs which passed yet didn't bring better params
    best_valid_loss = float('inf')
    best_valid_error = float('inf')
    
    
    sample_size, _ = train_X.shape
    batches = int(np.ceil(sample_size / mlp.batch_size))
    losses = [[], []] # losses[0] is for trainng, 1 for valid
    errors = [[], []] # [0] for training, 1 for valid
    
    for epoch in range(mlp.epochs):
        train_loss = 0.0
        for j in range(batches):
            b_start = j * mlp.batch_size
            b_end = min(sample_size, (j+1) * mlp.batch_size)
            batch_X = X[b_start:b_end, :]
            batch_y = y[b_start:b_end]
            grads = mlp._init_grads()
            for k in range(b_end - b_start):
                cache = mlp._fprop_one_example(batch_X[k])
                grads_tmp = mlp._bprop_one_example(cache, batch_y[k])
                for key in grads:
                    # compute mean grads[key]
                    grads[key] += (grads_tmp[key] - grads[key])/(k + 1)
                train_loss += mlp._compute_one_loss(cache["os"], batch_y[k])
                mlp._update_params(grads)
            # end of mini_batch

        train_loss /= sample_size 
        train_loss += mlp._compute_regular_loss()
        losses[0].append(train_loss)
        
        if early_stop == True:
            train_os = mlp.predict_probs(train_X)
            train_oy = np.argmax(train_os, axis = 1)
            train_error = mlp.compute_error(train_oy, train_y)
            errors[0].append(train_error)
            
            valid_os = mlp.predict_probs(valid_X)
            valid_oy = np.argmax(valid_os, axis = 1)
            valid_loss = mlp.compute_empirical_risk(valid_os, valid_y)
            losses[1].append(valid_loss)
            valid_error = mlp.compute_error(valid_oy, valid_y)
            errors[1].append(valid_error)
            
            if epoch == 0: # first epoch completed
                best_params = mlp._backup_params()
                best_valid_error = valid_error
            else:
                if valid_error > best_valid_error: # 
                    n_epochs_bad += 1
                    if n_epochs_bad == n_patientce: # stop
                        mlp._restore_params(best_params)
                        return mlp, losses, errors, best_valid_error
                else: # loss is decressing
                    best_params = mlp._backup_params()
                    best_valid_error = valid_error
                    n_epochs_bad = 0
    # end of one epoch
    return mlp, losses, errors, best_valid_error
````

````python
# hyper parameters selected.
n_hiddens = [4, 8]
early_stops = [True, False]
weight_decays = [[0.0]*4, [0.01]*4]

train_valid_X = np.vstack((train_X, valid_X))
train_valid_y = np.vstack((train_y.reshape(-1,1), valid_y.reshape(-1,1)))
train_valid = np.hstack((train_valid_X, train_valid_y))

min_x1, max_x1 = min(train_valid[:,0]), max(train_valid[:,0])
min_x2, max_x2 = min(train_valid[:,1]), max(train_valid[:,1])

n_points = 50
grid_x1 = np.linspace(min_x1, max_x1, num = n_points)
grid_x2 = np.linspace(min_x2, max_x2, num = n_points)
X1, X2 = np.meshgrid(np.arange(min_x1, max_x1, 0.02),
                     np.arange(min_x2, max_x2, 0.02))

grid_data = np.hstack((X1.reshape(-1, 1), X2.reshape(-1, 1)))
len_i = len(n_hiddens)
len_j = len(early_stops)
len_k = len(weight_decays)

plt.figure(figsize=(20, 12))
for i in range(len_i):
    for j in range(len_j):
        for k in range(len_k):
            mlp, _, _, _ = train_by_looping(train_X, train_y, 
                                         valid_X, valid_y,
                                         n_hiddens[i], 
                                         early_stops[j], 
                                         weight_decays[k],
                                         epochs = 2000
                                        )
            predicts = mlp.predict(grid_data).reshape(X1.shape)
            plt.subplot(len_i, len_j * len_k, 
                        i * (len_j * len_k) + (j * len_k) + k + 1)
            # plt.scatter(grid_data[:,0], grid_data[:,1], c = predicts, s=10)
            plt.imshow(predicts, extent=(min_x1, max_x1, min_x2, max_x2),
                       alpha=.4, origin='lower')
            plt.scatter(train_X[:,0], train_X[:,1], c = train_y, 
                        marker = 'v', s=10) # The training points
            plt.scatter(valid_X[:,0], valid_X[:,1], c = valid_y, 
                        marker = 's', s=10) # The test points
            plt.title("n_hidden:{}\n early_stop:{}\n weight_decay:{}"
                     .format(n_hiddens[i], early_stops[j], weight_decays[k]))
plt.show()            
````

````python
def train(train_X, train_y, valid_X, valid_y, test_X, test_y,
          n_hidden = 100, early_stop = False, weight_decay = [0.0]*4,
          learning_rate = 1e-3, epochs = 2000, 
          batch_size = 64, n_patience = 10,
          log_file_name = "mlp_training_log"):
    
    mlp = MLP(n_hidden = n_hidden,
              learning_rate = learning_rate,
              epochs = epochs,
              batch_size = batch_size,
              early_stop = early_stop,
              lambdas = weight_decay)
    
    mlp._init_weights_biases(train_X, train_y)
    # for early stopping
    best_params = None # params that performs best in valid set
    # n_patientce = 20 # maximun numbers of epoch tolerated when mlp
                     # with current params doesn't perform better than
                     # with best_params.
    n_epochs_bad = 0 # epochs which passed yet didn't bring better params
    best_valid_error = float('inf')
    
    sample_size, _ = train_X.shape
    batches = int(np.ceil(sample_size / mlp.batch_size))
    losses = [[], [], []] # [[train], [valid], [test]]
    errors = [[], [], []] # [[train], [valid], [test]]
    
    ISOTIMEFORMAT = '%Y-%m-%d %H:%M:%S'
    file_name = log_file_name + str(datetime.now().strftime(ISOTIMEFORMAT))
    file_name += ".txt"
    f = open(file_name, "w")
    f.write("training records of a MLP network\n")
    f.close()
    for epoch in range(mlp.epochs):
    # for epoch in range(mlp.epochs):
        # batch matrix forward and backward 
        for j in range(batches):
            batch_start = j * mlp.batch_size
            batch_end = min(sample_size, (j+1) * mlp.batch_size)
            batch_X = train_X[batch_start:batch_end, :]
            batch_y = train_y[batch_start:batch_end]
            cache = mlp._fprop(batch_X)
            grads = mlp._bprop(cache, batch_y)
            mlp._update_params(grads)
        # end of mini_batch

        # compute loss and error rate on train, valid, and test dataset
        dataset = [(train_X, train_y), (valid_X, valid_y), (test_X, test_y)]
        for i in range(len(dataset)):
            os = mlp.predict_probs(dataset[i][0])
            loss = mlp.compute_empirical_risk(os, dataset[i][1])
            losses[i].append(loss)
            oy = np.argmax(os, axis = 1)
            error = mlp.compute_error(oy, dataset[i][1])
            errors[i].append(error)
        
        # write to log file for current epoch
        f = open(file_name, "a+")
        log = "epoch {:<5}: ".format(epoch)
        log += "train_l:{:5.3f}, train_er:{:.3f}; ".format(
            losses[0][-1], errors[0][-1])
        log += "valid_l:{:5.3f}, valid_er:{:.3f}; ".format(
            losses[1][-1], errors[1][-1])
        log += "test_l:{:5.3f}, test_er:{:.3f}; ".format(
            losses[2][-1], errors[2][-1])
        log += "\n"
        f.write(log)
        f.close()
        
        # early stopping check and related process
        if early_stop == True:
            if epoch == 0: # first epoch completed
                best_params = mlp._backup_params()
                best_valid_error = errors[1][-1]  # using error 
            else:
                latest_valid_error = errors[1][-1]  # using error 
                if latest_valid_error > best_valid_error: # use > instead of >=
                    n_epochs_bad += 1
                    if n_epochs_bad == n_patience: # stop
                        mlp._restore_params(best_params)
                        best_epoch = epoch - n_patience
                        f = open(file_name, "a+")
                        log = "Early stop at epoch {}. With".format(best_epoch)
                        log += " valid_loss:{:.3f}, valid_error:{:.3f}".format(
                            losses[1][best_epoch], errors[1][best_epoch])
                        log += " train_loss:{:.3f}, train_error:{:.3f}".format(
                            losses[0][best_epoch], errors[0][best_epoch])
                        f.write(log)
                        f.close()
                        debug(log)
                        debug("log file: {}".format(file_name))
                        return mlp, losses, errors, best_valid_error
                else: # loss is decressing
                    best_params = mlp._backup_params()
                    best_valid_error = latest_valid_error
                    n_epochs_bad = 0
    # end of one epoch
    debug("training complete. log file: {}".format(file_name))
    f = open(file_name, "a+")
    log = "n_hidden:{:<5}\t".format(n_hidden)
    log += "learning_rate:{:.3f}\t".format(learning_rate)
    log += "batch_size:{:<3}\t".format(batch_size)
    log += "n_patience:{:<3}\t".format(n_patience)
    log += "weight_decay:{}\n".format(weight_decay)
    f.write(log)
    f.close()
    return mlp, losses, errors, best_valid_error


def plot_learning_curves(losses, errors, data_set_name = ""):
    support_x = np.arange(len(losses[0]))
    colors = ['g-','b-','r']
    set_groups = ["train_set", "valid_set", "test_set"]

    plt.figure(figsize=(13, 6))
    plt.grid(True) # add a grid

    plt.subplot(1, 2, 1)
    plt.title("Loss curves on {} dataset".format(data_set_name))
    legends = []
    for i in range(3):
        plt.xlabel("epoch")
        plt.ylabel("loss")
        plt.plot(support_x, losses[i], colors[i])
        legends.append(set_groups[i])
    plt.legend(legends)

    plt.subplot(1, 2, 2)
    plt.title("Error curves on {} dataset".format(data_set_name))
    for i in range(3):
        plt.xlabel("epoch")
        plt.ylabel("error")
        plt.plot(support_x, errors[i], colors[i])
    plt.legend(legends)
    plt.show()
````
